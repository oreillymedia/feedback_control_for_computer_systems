```python
# Input parameters
r = float( 4 )   # Reference or "setpoint"
k = float( 1.4 ) # Controller gain
u = 0            # "Previous" output
```

```python
for _ in range( 200 ):
    y = u        # One-step delay: previous output
    
    e = r - y    # Tracking error
    u = k*e      # Controller ouput

    print(r, e, 0, u, y)
```
